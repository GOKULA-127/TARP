```python
data={
    'gender':['male','female','male','male','female'],
    'grade':['a','b','a','b', 'a'],
    'score':[80,90,80,81,82]
}
df=pd.DataFrame(data)
print(df)

agg={
    'score':['mean','max']
}

group=df.groupby(['gender','grade']).aggregate(agg)
print(group)
```

```python
data = {
    'Category': ['A', 'A', 'B', 'B', 'A', 'B'],
    'Value': [10, 15, 20, 25, 30, 35]
}
df = pd.DataFrame(data)

agg={
    'Value':['mean','sum','max','min']
}

result=df.groupby(['Category']).aggregate(agg)
print(result)

des=df.describe()
print(des)
```

```python
import pandas as pd
data={
    'Region': ['North', 'South', 'North', 'South', 'East', 'East'],
    'Category': ['A', 'A', 'B', 'B', 'A', 'B'],
    'Sales': [1200, 800, 1500, 700, 900, 1100]
}

sales=pd.DataFrame(data)
print(sales)

summary=sales.pivot_table(
    values='Sales',
    index='Region',
    columns='Category',
    aggfunc=['mean','sum']
)
print(summary)
    
```

```python
employees = pd.DataFrame({
    'emp_id': [1, 2, 3],
    'name': ['Alice', 'Bob', 'Charlie']
})

salaries = pd.DataFrame({
    'emp_id': [2, 3, 4],
    'salary': [60000, 75000, 50000]
})

merged = pd.merge(employees, salaries, on='emp_id', how='left')
print(merged)
```

```python
df = pd.DataFrame({
    'city': ['Delhi', 'Mumbai', 'Delhi', 'Mumbai', 'Delhi', 'Mumbai'],
    'temperature': [30, 35, 32, 36, 29, 34]
})
result = df.groupby('city')['temperature'].agg(['mean', 'std'])
print(result)
```

```python
df = pd.DataFrame({
    'City': ['Delhi', 'Chennai', 'Bangalore', 'Mumbai'],
    'Temperature': [35, 38, 30, 33],
    'Humidity': [55, 70, 80, 60]
})

avg_temp = df.loc[df['Humidity'] > 60, 'Temperature'].mean()
print("Average Temperature where humidity > 60:", avg_temp)
```
